## what if the resource is not avaiable?

For example:
- song id 1353194608 is available
- song id 174124 is not available (2020/5/17)

### Solution 1: Use allow_redirects=False
However, even if the song is available, you still need redirections to get downloadable resource
`requests.get('http://music.163.com/song/media/outer/url?id=174124.mp3',allow_redirects=False)`

### Solution 2: Get final Url
Since
`requests.get('http://music.163.com/song/media/outer/url?id=174124.mp3',allow_redirects=False)` will redirect to 404 page, we can by getting final URL to check whether this song is available
You can use 

```python
r=requests.get('http://music.163.com/api/playlist/detail?id='+str(pid))
r.url()
```
or
```python
r=urllib.request.urlretrieve('link')
r.geturl()
```
After testing, the second code block works. The first sometime can't return the final url.

## Why you need a header?
1. if you didn't specify the header, sometime you can't really dowload a song.
the second song "Lost in paradise" is an example. With undownloaded mp3 file, the audiofile will be loaded as None and None has no tag attribution so you will get an error.
2. add header will boost the downloading process time.

````python
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)
````

## a stable way to create a new folder

````python
Path('./music').mkdir(parents=True, exist_ok=True)# an robust way to add a new folder
mp3_path='./music/'
Path('./img').mkdir(parents=True,exist_ok=True)
img_path='./img/'
````

## Be careful about the resource name

All resources reference name must not include '/'.
For album **Fate/Stay night A.OST**, it will pop up an error
`FileNotFoundError: [Errno 2] No such file or directory: './img/Fate/Stay night A.OST.jpg'`


So you should use
`        img_name=img_path+albums[i].replace('/','-')+'.jpg'`

For robustion, you should use
`        mp3_name=mp3_path+titles[i].replace('/','-')+'.mp3'`

## Main Function

````python


for i in tqdm(range(len(ids))):
    r=urllib.request.urlopen('http://music.163.com/song/media/outer/url?id='+str(ids[i])+'.mp3')
    if r.geturl() != 'https://music.163.com/404':
        mp3_name=mp3_path+titles[i].replace('/','-')+'.mp3'
        urllib.request.urlretrieve('http://music.163.com/song/media/outer/url?id='+str(ids[i])+'.mp3',mp3_name)
        
        img_name=img_path+albums[i].replace('/','-')+'.jpg'
        urllib.request.urlretrieve(album_pics[i],img_name)
        
        audiofile = eyed3.load(mp3_name)
        audiofile.tag.artist = str(artists[i])
        audiofile.tag.album = str(albums[i])
#         audiofile.tag.images.set(type_=3,img_data=None,mime_type='image/jpeg',img_url=album_pics[i])
# simple url reference won't work, you have to download it to your disk
# plus this method doesn't informed on the documentation, I found it on the stackoverflow
        audiofile.tag.images.set(type_=3,img_data=None,mime_type='image/jpeg',img_url=open(img_name,'rb').read())
# id3 version is important, encoding is important
# the former granteened the id3 tag will be recognized by music players like Apple Music
# the latter made the saving process ending up no error like 'Latin1' error
        audiofile.tag.save(version=eyed3.id3.ID3_V2_3,encoding='utf-8')
    
    
    
````

````python
import requests
import urllib.request
import eyed3
import os
from pathlib import Path
from tqdm.notebook import tqdm
import pandas as pd

def batch_download(pid):
    assert isinstance(pid,int),"id格式为纯数字"
    r=requests.get('http://music.163.com/api/playlist/detail?id='+str(pid))
    assert r.json()['code']==200,"未找到歌单"
    meta=r.json()['result']
    print('更新来自【'+ meta['creator']['nickname']+'】的歌单--'+meta['name'])
    ids=[]
    titles=[]
    artists=[]
    albums=[]
    album_pics=[]

    for i in range(len(r.json()['result']['tracks'])):
        ids.append(r.json()['result']['tracks'][i]['id'])
        titles.append(r.json()['result']['tracks'][i]['name'])
        artists.append(r.json()['result']['tracks'][i]['artists'][0]['name'])
        albums.append(r.json()['result']['tracks'][i]['album']['name'])
        album_pics.append(r.json()['result']['tracks'][i]['album']['blurPicUrl'])

    opener = urllib.request.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0')]
    urllib.request.install_opener(opener)
    Path('./music').mkdir(parents=True, exist_ok=True)# an robust way to add a new folder
    mp3_path='./music/'
    Path('./img').mkdir(parents=True,exist_ok=True)
    img_path='./img/'

    d={ 'ids':ids,'titles': titles, 'artists': artists, 'albums': albums, 'album_pics':album_pics}
    df=pd.DataFrame(data=d)

    old = [os.path.splitext(item)[0] for item in os.listdir(mp3_path)]
    new = list(set(titles)-set(old))
    print('本次更新曲目',*new,sep=',')
    for i in tqdm(range(0,len(new))):
    #     row=df.loc[df.titles == new[i]]
        #reference cell value row.titles.values[0]
    # better way
        row=df.loc[df.titles == new[i]].to_dict(orient='records')[0]
        song_url='http://music.163.com/song/media/outer/url?id='+str(row['ids'])+'.mp3'
        r=urllib.request.urlopen(song_url)
        if r.geturl() == 'https://music.163.com/404':
            print('歌曲【'+row['titles']+"】无资源！")
        else:
            mp3_name=mp3_path+row['titles'].replace('/','-')+'.mp3'
            urllib.request.urlretrieve(song_url,mp3_name)

            img_name=img_path+row['albums'].replace('/','-')+'.jpg'
            urllib.request.urlretrieve(row['album_pics'],img_name)

            audiofile = eyed3.load(mp3_name)
            audiofile.tag.artist = str(row['artists'])
            audiofile.tag.album = str(row['albums'])
    #         audiofile.tag.images.set(type_=3,img_data=None,mime_type='image/jpeg',img_url=album_pics[i])
    # simple url reference won't work, you have to download it to your disk
    # plus this method doesn't informed on the documentation, I found it on the stackoverflow
            audiofile.tag.images.set(type_=3,img_data=None,mime_type='image/jpeg',img_url=open(img_name,'rb').read())
    # id3 version is important, encoding is important
    # the former granteened the id3 tag will be recognized by music players like Apple Music
    # the latter made the saving process ending up no error like 'Latin1' error
            audiofile.tag.save(version=eyed3.id3.ID3_V2_3,encoding='utf-8')
````
